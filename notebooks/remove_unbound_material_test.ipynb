### Purpose

Initialize test directory which adds extra_heat and then attempts to implement a wind which removes unbound material. Constructing this file will initially be done by hand a bit.

```python
from qol.mesa.launcher import *
import qol.info as info

run_path = f'{info.qol_path}/notebooks/explode_sun_test/'
```

```python
from qol.mesa.launcher import *
import qol.info as info

# Make inlist
inlist = MesaInlist('inlist_explode_sun', LOGS_dir='LOGS/explode_sun/', photos_dir='photos/explode_sun/')
inlist.enable_pgstar()
inlist.save_pgstar(write_path='Grid1/evolve_rg/')
inlist.use_qol_pgstar()

inlist.enable_hydrodynamics()
inlist.initial_mass(1.)
inlist.set_Zbase(0.02)
inlist.disable_nuclear_burning()

# Make working directory
work = MesaWorkingDirectory(run_path=run_path)
work.copy_history_columns_list(f'{info.qol_path}mesa/resources/r24.08.1/history_columns.list')
work.copy_profile_columns_list(f'{info.qol_path}mesa/resources/r24.08.1/profile_columns.list')
work.load_qol_pgstar()

work.add_task(inlist)

work.save_directory(grant_perms=True)
```
