Purpose: To test the new functionality I put into the pyplot wrapper.

```python
import qol.vis.pyplot as plt
import numpy as np
```

```python
ax = plt.axes()

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.set_title('title')

x = np.linspace(0.01, 1, 100)

ax.plot(x, x**2)
ax.plot(x, x**3)
ax.plot(x, x**4)
ax.plot(x, x**2+0.2)

ax.scatter(x, x**2)
ax.scatter(x, x**3)
ax.scatter(x, x**4, edgecolor='k')
ax.scatter(x, x**2+0.2, edgecolor='k', zorder=5, lw=10)

plt.loglog()

# plt.axes()

plt.show()
```
